<span style="font-family:Calibri; font-size:1.2em">Data Manipulation steps to be carried out:</span>
 
 1. Convert ActivityDate to datetime64 dtype and format as yyyy-mm-dd
 2. Create a new column <mark>DayOfWeek</mark> to track the day of the week for improved analysis
 3. Create a new column <mark>TotalMinutes</mark> = VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes + SedentaryMinutes
 4. Create a new column <mark>TotalHours</mark> by converting TotalMinutes column
 5. Rearrange and rename a few columns
 
<span style="font-family:Times New Roman; font-size:1.2em">Converting ActivityDate to dtype datetime64 and changing the format to yyyy-mm-dd</span>

```python
# converting "ActivityDate" to datatime64 dtype and format to dd-mm-yyyy
daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'], format = "%m/%d/%Y")

daily_activity.info() # Verifying whether the conversion of "ActivityDate" has been successful

daily_activity['ActivityDate'].head() # First 5 rows of ActivityDate to confirm the format change
```

<span style="font-family:Times New Roman; font-size:1.2em">Creating new list with by rearranging column names and using df_dact instead of daily_activity</span>

```python
# new list of rearranged columns
new_cols = ['Id', 'ActivityDate', 'DayOfWeek', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalMinutes', 'TotalHours', 'Calories']

# changing columns to new_cols list
df_dact = daily_activity.reindex(columns=new_cols)

# verifying whether the change is successful by printing first 5 rows
df_dact.head(5)
```

<span style="font-family:Times New Roman; font-size:1.2em">Creating new column "DayOfWeek" to separate the date to extract the day of the week for further analysis of the dataset</span>

```python
df_dact['DayOfWeek'] = df_dact['ActivityDate'].dt.day_name() # DayOfWeek to represent day of the week

df_dact['DayOfWeek'].head(5)
```

<span style="font-family:Times New Roman; font-size:1.2em">Creating new column "TotalMinutes" to log the total time</span>

```python
# TotalMinutes column created by adding up all the minutes
df_dact['TotalMinutes'] = df_dact['VeryActiveMinutes'] + df_dact['FairlyActiveMinutes'] + df_dact['LightlyActiveMinutes'] + df_dact['SedentaryMinutes']

df_dact['TotalMinutes'].head(5)
```

<span style="font-family:Times New Roman; font-size:1.2em">Creating new column "TotalHours" by converting the values in "TotalMinutes" column to hours</span>

```python
# TotalHours column created by converting TotalMinutes to hour and rounding the values
df_dact['TotalHours'] = round(df_dact['TotalMinutes'] / 60)

df_dact['TotalHours'].head(5)
```

<span style="font-family:Calibri; font-size:1.3em">Data is now ready to be analyzed</span>

#### <span style="font-family:Calibri; font-size:2em; color:orange">STEP 5: SHARE</span>

<span style="font-family:Calibri; font-size:1.2em;">5.1 Data Visualization</span>

```python
# plotting histogram to analyze the no. of times users logged in across the week

plt.figure(figsize=(9,8)) # size of the chart
plt.hist(df_dact['DayOfWeek'], bins = 7, 
         width = 0.6, color = "lightblue", edgecolor = "black")

# annotations and visuals 
plt.xlabel("Day of the week")
plt.ylabel("Frequency")
plt.title("No. of times users logged in app through the week")
plt.tight_layout()
```

<span style="font-family:Calibri; font-size:1.2em;">Observations from the frequency plot:-</span>

1. Users prefer or use to log their activity status mid-week that is from Tuesday to Friday.
2. Frequency of usage drops on Friday and then continues to be same till Monday.

```python
plt.figure(figsize=(10,8)) # specifying the size of the plot
plt.scatter(df_dact['TotalSteps'], df_dact['Calories'],
           alpha = 0.5, c = df_dact['Calories'],
           cmap = 'coolwarm')

mean_calories = 2303
mean_steps = 7637

plt.colorbar(orientation = "vertical")
plt.axvline(mean_steps, color = "Blue", label = "Mean steps")
plt.axhline(mean_calories, color = "Red", label = "Mean calories burned")
plt.xlabel("Steps taken")
plt.ylabel("Calories burned")
plt.title("Calories burned for every step taken")
plt.legend(loc='lower right')
plt.tight_layout()
```

<span style="font-family:Calibri; font-size:1.2em;">Observations from plot:-</span>

1. Calories burned for every step taken is a positive correlation
2. Intensity of calories burned increased where users have logged steps from 0 to 15,000 and cooling down with steps < 15,000
3. A few more noticeable observations:-
    - One observation of steps > 35,000 with < 3,000 calories burned (Not in accordance with the rest of the data)
    - Zero steps having 0 to atleast 500 calories burned as observed
    - Deducing if outliers might be due to errors in data collection (miscalculation, data contamination, human errors or faulty device)

```python
plt.figure(figsize=(10,8)) # specifying the size of the plot
plt.scatter(df_dact['TotalHours'], df_dact['Calories'],
           alpha = 0.8, c = df_dact['Calories'],
           cmap = 'coolwarm')

mean_calories = 2303
mean_total_hours = 20
mean_sedentary_hours = 991 / 60

plt.colorbar(orientation = "vertical")
plt.axvline(mean_total_hours, color = "Blue", label = "Mean total hours")
plt.axvline(mean_sedentary_hours, color = "Yellow", label = "Mean sedentary hours")
plt.axhline(mean_calories, color = "Red", label = "Mean calories burned")
plt.xlabel("Hours")
plt.ylabel("Calories burned")
plt.title("Calories burned for every step taken")
plt.legend(loc='upper left')
plt.tight_layout()
```

<span style="font-family:Calibri; font-size:1.2em;">Observations from the Calories burned for every step taken plot:-</span>

1. A weak positive correlation whereby the increase of hours logged does not translate to more hours being logged. That is largely due to the average sedentary hours (yellow line) plotted 
2. Again we can see a few outliers:-
    - Same zero value outliers as calories burned every step scatter plot
    - A blue dot at the 24 hour mark with 0 calorie burned

```python
# calculating total of individual minutes columns
very_active_minutes = df_dact['VeryActiveMinutes'].sum()
fairly_active_minutes = df_dact['FairlyActiveMinutes'].sum()
lightly_active_minutes = df_dact['LightlyActiveMinutes'].sum()
sedentary_minutes = df_dact['SedentaryMinutes'].sum()

# plotting a pie chart
plt.figure(figsize=(9,7))
pie_slices = [very_active_minutes, fairly_active_minutes, lightly_active_minutes, sedentary_minutes]
slice_labels = ['very_active_minutes', 'fairly_active_minutes', 'lightly_active_minutes', 'sedentary_minutes']
slice_color = ['lightgreen', 'lightskyblue', 'coral', 'pink']
explode = [0,0,0,0.1] # exploding sedentary_minutes from the pie chart
plt.pie(pie_slices, labels = slice_labels,
       colors = slice_color, wedgeprops = {'edgecolor':'black'},
       explode = explode, autopct = "%.1f%%")
plt.title('Activity Percentage (In Minutes)')
plt.tight_layout()
```

<span style="font-family:Calibri; font-size:1.2em"></span>

<span style="font-family:Calibri; font-size:1.2em;">Observations from the Activity Percentage (In Minutes) pie chart:-</span>

1. Sedentary minutes covers the maximum percentage of activity = 81.3%
2. The observation concludes that FitBit app users are logging daily activites for inactive movements, daily commute, and activities which involve less movement
3. App is not being used to track fitness activities (running, exercising) as can be seen from the minute percentages - 1.1% for fairly_active_minutes and 1.7% for very_active_minutes. This proves that the FitBit app does not serve its purpose, which is to encourage people to stay fit.
